```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
df = pd.read_csv('/content/world_population.csv')
df
```

```python
# 10 most populated countries
df[df['Rank'] <= 10].sort_values(by = 'Rank')
```

```python
df = df.copy()
```

```python
#checking for missing values
missing_count = df.isnull().sum()
value_count = df.isnull().count()

missing_percentage = round(missing_count/ value_count * 100, 2)
missing_df = pd.DataFrame({'count': missing_count, 'percentage': value_count})
missing_df
```

```python
# filling missing values
df = df.fillna(method= 'pad')
df
```

```python
df.isnull().sum()
```

```python
# checking for duplicates
def duplicated_values(df):
  print(f'duplicate check...')
  duplicated = df.duplicated(subset ='Country' , keep = 'first').sum()
  if duplicated > 0:
    print(f'there are {duplicated} observations')
    df.drop_duplicates(keep= 'first', inplace= True)
    print(f'{duplicated} were dropped')
    print(f'no more duplicates')
  else:
      print(f'there are no duplicates')
duplicated_values(df)
```
